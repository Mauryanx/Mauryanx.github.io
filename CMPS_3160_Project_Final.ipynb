Model Questions: 

Using our first dataset we will plan to identify players that are making significant contributions but are not getting enough playing time. 

This is very important as we can then identify good players who will also be willing to transfer as they might feel like their talents are not being fully used.

Using our second dataset , we can then crossvalue the these players with their age , BMI and pace and wage  to see which  players would be a best pick.
To put this in question form.

"We plan to see how managers can use analystics to optimize their transfers?"

Another question we can plan to answer is from this article "https://www.espn.com/soccer/english-premier-league/23/blog/post/2619206/pace-is-the-premier-leagues-fundamental-attribute"

This article states that pace is the most fundamental and therefore significant attribute of a player. We can use our first dataset of a players contributions to a game and check its correlation with their pace to see whether this hypothesis can be backed by statistics.

To put this in question form;

" Is pace the most significant attribute, and how important are the other attributes of a soccer player ?

We therefore built a KNN regressors model , we incorporated proper data science techniques and further built a method to find the optimal number of neighbours for our model.We decided to only model for attackers and therefore made the reasoned assumption that attributes not valid for strikers would negatively affect our model and thus made the decision to omit those.We further calculated an estimate of test error based on 10-fold cross validation.

```python
# What does FIFA think is the most important attributes for a player?
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

features = ["age", "height_cm", "weight_kg", "nationality", "preferred_foot", "weak_foot", "skill_moves", "work_rate", "body_type", "pace", "shooting", "passing", "dribbling", "defending", "physic", "attacking_crossing", "attacking_finishing", "attacking_heading_accuracy", "attacking_short_passing", "attacking_volleys", "skill_dribbling", "skill_curve", "skill_fk_accuracy", "skill_long_passing", "skill_ball_control", "movement_acceleration", "movement_sprint_speed", "movement_agility", "movement_reactions", "movement_balance", "power_shot_power", "power_jumping", "power_stamina", "power_strength", "power_long_shots", "mentality_aggression", "mentality_interceptions", "mentality_positioning", "mentality_vision", "mentality_penalties", "mentality_composure", "bmi"]
X_train_dict = most_valuable_strikers[features].to_dict(orient="records")
y_train = most_valuable_strikers["overall"]

vec = DictVectorizer(sparse=False)
scaler = StandardScaler()
model = KNeighborsRegressor(n_neighbors=4)
pipeline = Pipeline([("vectorizer", vec), ("scaler", scaler), ("fit", model)])

np.mean(
    -cross_val_score(pipeline, X_train_dict, y_train, cv=10, scoring="neg_mean_squared_error")
)
```

```python
from sklearn.model_selection import cross_val_score
# calculates estimate of test error based on 10-fold cross validation
def get_cv_error(k):
    model = KNeighborsRegressor(n_neighbors=k)
    pipeline = Pipeline([("vectorizer", vec), ("scaler", scaler), ("fit", model)])
    mse = np.mean(-cross_val_score(
        pipeline, X_train_dict, y_train, 
        cv=10, scoring="neg_mean_squared_error"
    ))
    return mse
    
ks = pd.Series(range(1, 51))
ks.index = range(1, 51)
test_errs = ks.apply(get_cv_error)

test_errs.plot.line()
test_errs.sort_values()
```

```python
from sklearn.model_selection import cross_val_score
# calculates estimate of test error based on 10-fold cross validation
def get_cv_error(k):
    model = KNeighborsRegressor(n_neighbors=k)
    pipeline = Pipeline([("vectorizer", vec), ("scaler", scaler), ("fit", model)])
    mse = np.mean(-cross_val_score(
        pipeline, X_train_dict, y_train, 
        cv=10, scoring="neg_mean_squared_error"
    ))
    return mse
    
ks = pd.Series(range(1, 51))
ks.index = range(1, 51)
test_errs = ks.apply(get_cv_error)

test_errs.plot.line()
test_errs.sort_values()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train_dict, y_train, test_size=0.2)

pipeline = Pipeline([
    ("vectorizer", vec), ("scaler", scaler),
    ("model", LinearRegression())
])
pipeline.fit(X_train, y_train)

predictions = pipeline.predict(X_test)
```

```python
# Access the coefficients and feature names
coefficients = pipeline.steps[2][1].coef_
feature_names = pipeline.steps[0][1].get_feature_names()

# Create a bar plot
plt.bar(feature_names, coefficients)
plt.xlabel("Feature names")
plt.ylabel("Coefficients")
plt.title("Feature importance")

# Rotate the x-axis tick labels by 45 degrees
plt.xticks(rotation=45)

# Add a legend
plt.legend(["Coefficient"])

plt.show()
```

This is our final bar chart we got from our data , which visually shows the importance of each attribute in relation to each other and their respective coefficients. We see that shooting is most significant which is likely as strikers inherently need to have good shooting to convert shots into goals.

```python
# Zip the coefficients and feature names together
coefs_with_names = zip(coefficients, feature_names)

# Sort the coefficients in descending order
sorted_coefs = sorted(coefs_with_names, key=lambda x: x[0], reverse=True)

# Print the coefficients and their corresponding feature names
for coef, name in sorted_coefs:
    print(f"{name}: {coef}")
```
